```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# ========== CONFIG ==========
API_KEY = "2e879012319b4f"   # Replace with your OpenWeatherMap API Key
CITY = "Chennai"
URL = f"https://api.openweathermap.org/data/2.5/forecast?q={CITY}&appid={API_KEY}&units=metric"

# ========== FETCH DATA ==========
response = requests.get(URL)
data = response.json()

forecast_list = data["list"]

dates = []
temps = []

for item in forecast_list:
    dates.append(item["dt_txt"])
    temps.append(item["main"]["temp"])

df = pd.DataFrame({"Date_Time": dates, "Temperature (°C)": temps})
df["Date_Time"] = pd.to_datetime(df["Date_Time"])  # convert for better plotting

# ========== VISUALIZE ==========
plt.figure(figsize=(12,5))
plt.plot(df["Date_Time"], df["Temperature (°C)"], marker="o")
plt.xticks(rotation=45)
plt.title(f"5-Day Temperature Forecast - {CITY}")
plt.xlabel("Date & Time")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
